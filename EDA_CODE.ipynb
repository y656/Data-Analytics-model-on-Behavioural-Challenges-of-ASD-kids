### Handling outliers 
Outliers are data which are on the extreme sides(low or high). Handling outliers involve:
<ul>
    <li>Detecting outliers</li>
    <li>Treatment of outliers</li>
</ul>

For this consider any column and determine the upper cutoff and lower cutoff with the help of any of 3 methods namely:
<ul>
    <li>Percentile Method</li>
    <li>IQR Method</li>
    <li>Standard Deviation Method</li>
</ul>

We will be determining if there are outliers in our data using IQR(Inter Quartile Range) method. 
Lets select Social_Responsiveness_Scale as variable to find outliers.


```python
p0 = df.Social_Responsiveness_Scale.min()
p100 = df.Social_Responsiveness_Scale.max()
q1 = df.Social_Responsiveness_Scale.quantile(0.25)
q2 = df.Social_Responsiveness_Scale.quantile(0.5)
q3 = df.Social_Responsiveness_Scale.quantile(0.75)
iqr = q3-q1
```

```python
lc = q1 - 1.5*iqr
uc = q3 + 1.5*iqr
```

If lc < p0 → There are NO Outliers on the lower side

If uc > p100 → There are NO Outliers on the higher side

```python
print( "p0 = " , p0 ,", p100 = " , p100 ,", lc = " , lc ,", uc = " , uc)
```

We can observe that both the conditions are satisfied by our dataset. So, there are no outliers present in our dataset.

We can also get the pictorial representation and detect any outliers present using Box Plot

```python
df.Social_Responsiveness_Scale.plot(kind ='box')
```

### Missing values treatment

To treat missing values we can adopt one of the three methods:

<ul>
    <li>Drop the variable</li>
    <li>Drop the observation(s)</li>
    <li>Missing Value Imputation</li>
</ul>

As we have less percentage(0-5%) of missing values we can drop the observations. If we have percentage(like 5-40percentage and above) we can then do missing value imputation by using statistical measures like median, mean for numerical values and mode for categorical variables. If we have higher percentage like(50 percentage above) then drop that column

### Importing required libraries:

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier,AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from xgboost import XGBClassifier
from sklearn.preprocessing import OneHotEncoder,LabelEncoder
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold
from sklearn.compose import make_column_transformer
from sklearn import tree
from sklearn.model_selection import GridSearchCV
import xgboost as xgb
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
```

```python
X = df.drop('ASD_traits',axis =1)
y = df['ASD_traits']
```

One Hot encoding of categorical variables so we can use them in algorithms

```python
l = ['Who_completed_the_test','Family_mem_with_ASD','Jaundice','Ethnicity','Sex','Anxiety_disorder','Social/Behavioural Issues',
    'Global developmental delay/intellectual disability','Depression','Speech Delay/Language Disorder',
    'Learning disorder','Genetic_Disorders']
```

```python
transformer = make_column_transformer((OneHotEncoder(),l),remainder='passthrough')
transformed = transformer.fit_transform(X)
transformed_df = pd.DataFrame(transformed, columns=transformer.get_feature_names_out())
```

```python
transformed_df.head(5)
```

```python
en = LabelEncoder()
y= en.fit_transform(y)
y
```

Scaling the dataset to avoid unnecessary weightage using Standard scaler

```python
cols = transformed_df.columns
```

```python
transformed_df = StandardScaler().fit_transform(transformed_df)
```

### Decision Tree Classifier:

```python
model = DecisionTreeClassifier(criterion = 'gini')
model.fit(transformed_df,y)
```

Lets visualize how our decision tree looks like:)

```python
plt.figure(figsize =(20,40))
tree.plot_tree(model, feature_names = cols,class_names=['Disease', "No Disease"],filled = True,fontsize = 9)
plt.show()
```

```python
kf = StratifiedKFold(n_splits=10, shuffle=False)
scores = cross_val_score(model, transformed_df, y, cv=20)
scores
```

```python
avg_score = np.mean(scores)
print(avg_score)
```

Performing K-fold cross validation over 20 splits and averaging out scores resulted in 97.34% accuracy

### Random Forest Classifier:

```python
clf = RandomForestClassifier(n_estimators = 200,criterion = 'entropy') 
clf.fit(transformed_df, y)
y_pred = clf.predict(transformed_df)
print(classification_report(y_pred, y))
scores = cross_val_score(clf, transformed_df, y, cv=20)
scores
```

```python
avg_score = np.mean(scores)
print(avg_score)
```

We can see that we achieved  around 100 percentage accuracy using Random Forest Classifier

### Logistic Regression:

```python
clf = LogisticRegression(max_iter = 200,solver ='newton-cg')
clf.fit(transformed_df, y)
scores = cross_val_score(clf, transformed_df, y, cv=20)
avg_score = np.mean(scores)
print(avg_score)
```

We got 96.7 % accuracy using Logistic Regression

### Xgboost classifier:

```python
clf = XGBClassifier(learning_rate =0.0099,
 n_estimators=200,
 max_depth=6,
 min_child_weight=1,
 gamma=0,
 subsample=0.2,
 colsample_bytree=0.75,
 objective= 'binary:logistic',
 nthread=4,
 reg_alpha =1e-4,
 scale_pos_weight=1,
 seed=27)
clf.fit(transformed_df, y)
scores = cross_val_score(clf, transformed_df, y, cv=20)
scores
```

```python
avg_score = np.mean(scores)
print(avg_score)
```

We got 98.3 % accuracy using Xgboost Classifier

### Adaboost Classifier:

```python
clf = AdaBoostClassifier(n_estimators = 100,learning_rate = 1,algorithm = 'SAMME') 
clf.fit(transformed_df, y)
scores = cross_val_score(clf, transformed_df, y, cv=20)
scores
```

```python
avg_score = np.mean(scores)
print(avg_score)
```

After Hyperparameter tuning we got avg score of 95.6% using Adaboost algorithm

```python
param_grid = {
    'n_estimators':[40,50,100,120],
    'learning_rate':[0.7,0.8,1,2],
    'algorithm':['SAMME','SAMME.R']
}
 
grid = GridSearchCV(AdaBoostClassifier(), param_grid, refit = True, verbose = 3)
grid.fit(transformed_df, y)
print(grid.best_params_)
 
# print how our model looks after hyper-parameter tuning
print(grid.best_estimator_)
```

We achieved 95.6% accuracy using Adaboost classifier
